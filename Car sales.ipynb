```python
#data cleaning
data['Date']=pd.to_datetime(data['Date'])
```

```python
data['Year']=data['Date'].dt.year
data['Month']=data['Date'].dt.month
data['Month_Name']=data['Date'].dt.month_name()

```

```python
data['Engine']=data['Engine'].str.replace('Â','',regex=False)
```

```python
data['Company'].value_counts().head(10)
plt.figure(figsize=(10,5))
data['Company'].value_counts().head(10).plot(kind='bar')
plt.title('Toop 10 Car Companies by Sales')
```

```python
data.groupby('Company')['Price ($)'].sum().sort_values(ascending=False).head(10)
```

```python
data['Transmission'].value_counts()

sns.countplot(x='Transmission',data=data)
plt.title("Transmission Type Distribution")
plt.show()
```

```python
data.groupby('Body Style')['Price ($)'].mean().sort_values(ascending=False)
```

```python
sns.countplot(x="Gender",data=data)
plt.title("Gender Distribution")
plt.show()
```

```python
data.groupby('Dealer_Region')['Price ($)'].sum().sort_values(ascending=False)
```

```python
def price_category(price):
    if price < 20000:
        return 'Budget'
    elif price <= 30000:
        return 'Mid Range'
    else:
        return 'Premium'

data['Price_Category'] = data['Price ($)'].apply(price_category)
```
